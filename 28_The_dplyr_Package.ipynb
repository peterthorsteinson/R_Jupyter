## Grammar of Data Manipulation
Hadley Wickham’s package dplyr package is optimized to work efficiently with data frames. The dplyr package is a grammar of data manipulation that provides verbs that act on a data frame to perform data manipulation:
- **mutate()** adds or modifies variables as functions of other variables
- **select()** returns a subset of columns based on column names
- **filter()** extracts subset of rows based on logical conditions
- **summarise()** generates summary statistics
- **arrange()** reorders rows
- **sample_n()** take random sample
- **sample_frac()** take random sample
- **rename()** renames variables
- **%>%** pipe operator connects verb actions into a pipeline

## Common Characteristics of dplyr Function
- The first argument is a data frame
- The subsequent arguments describe what to do with the first argument
- The return value is a new data frame
- Data frames must be tidy (properly formatted and annotated)

## The starwars Dataset

```r
library(dplyr)
dim(starwars)    # dimensions of the data frame
names(starwars)  # columns names
starwars %>% 
  select(name, ends_with("color")) # select columns of a data frame that you want to focus on
```

```r
library(dplyr)
starwars %>% 
  mutate(name, bmi = mass / ((height / 100)  ^ 2)) %>%
  select(name:mass, bmi)
```

```r
library(dplyr)
starwars %>% 
  arrange(desc(mass))
```

```r
library(dplyr)
starwars %>%
  group_by(species) %>%
  summarise(
    n = n(),
    mass = mean(mass, na.rm = TRUE)
  ) %>%
  filter(n > 1)
```

## R for Data Science (Grolemund & Wickham): [Data transformation](http://r4ds.had.co.nz/transform.html)

Tibbles are data frames, but slightly tweaked to work better in the tidyverse.

```r
library(tidyverse)
library(nycflights13)

by_dest <- group_by(flights, dest)
delay <- summarise(by_dest,
  count = n(),
  dist = mean(distance, na.rm = TRUE),
  delay = mean(arr_delay, na.rm = TRUE)
)
delay <- filter(delay, count > 20, dest != "HNL")

# It looks like delays increase with distance up to ~750 miles 
# and then decrease. Maybe as flights get longer there's more 
# ability to make up delays in the air?
ggplot(data = delay, mapping = aes(x = dist, y = delay)) +
  geom_point(aes(size = count), alpha = 1/3) +
  geom_smooth(method = "loess")
```

```r
library(tidyverse)
library(nycflights13)

not_cancelled <- flights %>% 
  filter(!is.na(dep_delay), !is.na(arr_delay))

not_cancelled %>% 
  group_by(year, month, day) %>% 
  summarise(mean = mean(dep_delay))

delays <- not_cancelled %>% 
  group_by(tailnum) %>% 
  summarise(
    delay = mean(arr_delay, na.rm = TRUE),
    n = n()
  )

ggplot(data = delays, mapping = aes(x = n, y = delay)) + 
  geom_point(alpha = 1/10)

delays %>% 
  filter(n > 25) %>% 
  ggplot(mapping = aes(x = n, y = delay)) + 
    geom_point(alpha = 1/10)
```
