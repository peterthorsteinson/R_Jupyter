# 23 Linear Regression  
Linear regression is a very simple approach for supervised learning, for predicting a quantitative output.

[UC Business Analytics R Programming Guide: Linear Regression](http://uc-r.github.io/linear_regression)  

```r
# x <- c(176, 154, 138, 196, 132, 176, 181, 169, 150, 175)
# y <- c(82, 49, 53, 112, 47, 69, 77, 71, 62, 78)

x <- abs(rnorm(100))
y <- 3*x + rnorm(100) + 6

lmfit <- lm(y ~ x)
summary(lmfit)

plot(x, y)
abline(lmfit)
```

```r
count <- 20
β0 <- 3
β1 <- 0.5
ε <- rnorm(count)
ε
x <- c(1:count)
y <- β1*x + β0 + ε
y

plot(x, y, xlim=range(0:count), ylim=range(0:count))

curve(β1*x + β0, 0, count, add=TRUE)           # actual linear relationship

model <- lm(y ~ x)
m <- model$coef["x"]
b <- model$coef["(Intercept)"]
curve(m*x + b, 0, count, add=TRUE, col="red")  # modelled linear relationship
```
