# 29 The tibble Package

The tibble package is one of the unifying features of the tidyverse. It is a modern version of the legacy data.frame.

## Installation

The easiest way to get tibble is to install the whole tidyverse:  
**install.packages("tidyverse")**

Alternatively, install just tibble:  
**install.packages("tibble")**

Or the development version from GitHub:  
**install.packages("devtools")**  
**devtools::install_github("tidyverse/tibble")**  

[CRAN: tibble](https://cran.r-project.org/web/packages/tibble/README.html)  
[R for Data Science (Grolemund & Wickham) - Tibbles](http://r4ds.had.co.nz/tibbles.html) 
 


```r
# You can create a new tibble from avectors with the tibble() function.
library(tidyverse)
tibble(
  x = 1:5, 
  y = 1, 
  z = x ^ 2 + y
)
```

```r
# You can coerce a data frame into a tibble with the as_tibble() function.
library(tidyverse)
as_tibble(iris)
```

```r
# You can define a tibble row-by-row with tribble() function (note thr 'r' in 'tribble')
library(tidyverse)
tribble(
  ~x, ~y,  ~z,
  "a", 2,  3.6,
  "b", 1,  8.5
)
```
