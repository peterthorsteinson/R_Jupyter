## The plot Function

A scatterplot is displayed when you pass two numeric vectors as arguments to the plot() function.

```r
# the plot function
x <- c(0.5, 2.7, 4.2, 6.9, 11.8, 13)
y1 <- c(1, 3.3, 2.9, 3.4, 4.2, 3.8)
y2 <- c(4, 0.8, 0.55, 0, 5.4, 0.3)
par(las=1, mar=c(4, 4, 2, 4))
plot.new()
plot.window(range(x), c(0, 6))
lines(x, y1)
lines(x, y2)
points(x, y1, pch=16, cex=2)
points(x, y2, pch=21, bg="yellow", cex=2)
par(col="grey50", fg="grey50", col.axis="grey50")
axis(1, at=seq(0, 16, 4))
axis(2, at=seq(0, 6, 2))
axis(4, at=seq(0, 6, 2))
mtext("weight", side=1, line=2, cex=0.8)
mtext("score", side=2, line=2, las=0, cex=0.8)
mtext("speed", side=4, line=2, las=0, cex=0.8)
text(4, 5, "Great Spotted Kiwi")
par(mar=c(5.1, 4.1, 4.1, 2.1), col="black", fg="black", col.axis="black")
```

## The hist Function

The hist function takes in a vector of values that is displayed as a histogram.

```r
Temperature <- airquality$Temp
hist(Temperature)
```

```r
x <- read.csv("data/histogram.csv", header=T, sep="\t")
x <- t(x)
head(x)
y <- as.numeric(x[2, 1:ncol(x)])
hist(y, main="Histogram of histogram.csv")
```

# The boxplot Function

Boxplot: The boxplot function takes in any number of numeric vectors and draws a boxplot for each vector.

```r
boxplot(airquality$Ozone)
```

```r
par(mar=c(3, 4.1, 2, 0))
     boxplot(len ~ dose, data = ToothGrowth,
             boxwex = 0.25, at = 1:3 - 0.2,
             subset= supp == "VC", col="grey90",
             xlab="",
             ylab="tooth length", ylim=c(0,35))
     mtext("Vitamin C dose (mg)", side=1, line=2.5, cex=0.8)
     boxplot(len ~ dose, data = ToothGrowth, add = TRUE,
             boxwex = 0.25, at = 1:3 + 0.2,
             subset= supp == "OJ", col="grey70")
     legend(1.5, 9, c("Ascorbic acid", "Orange juice"), bty="n",
            fill = c("grey90", "grey70"))
par(mar=c(5.1, 4.1, 4.1, 2.1))
```

## The persp Function

The persp Function draws perspective plots of a surface over the x–y plane.

```r
x <- 0:9
x
y <- 0:9
y
f <- function(x, y) { x + y }
z <- outer(x, y, f)
z
persp (x, y, z)
```

```r
x <- seq(-10, 10, length= 30)
y <- x
f <- function(x, y) { r <- sqrt(x^2+y^2); 10 * sin(r)/r }
z <- outer(x, y, f)
persp(x, y, z, main = "Plot of sin(x) + cos(y)", xlab = "x axis", ylab = "y axis", zlab = "z axis", theta = 45, phi = 40)
image(x,y,z,col=terrain.colors(20),xlab="X",ylab="Y") 
```

```r
x <- seq(-10, 10, length= 30)
y <- x
f <- function(x,y) { r <- sqrt(x^2+y^2); 10 * sin(r)/r }
z <- outer(x, y, f)
z[is.na(z)] <- 1
# 0.5 to include z axis label
par(mar=c(0, 0.5, 0, 0), lwd=0.1)
     persp(x, y, z, theta = 30, phi = 30, expand = 0.5, col = "grey80")
par(mar=c(5.1, 4.1, 4.1, 2.1), lwd=1)
```

```r
saddle <- function(x, y){
 x^2-y^2
}

x <- y <- seq(-1, 1, length= 20)
z <- outer(x, y, saddle)

persp(x, y, z)
```

# The pie Function

The pie Function function takes in a vector of non-negative numbers that are displayed as the areas of pie slices.

```r
# simple pie chart
budget = c(Rent=800, Food=300, Clothing=150, Transportation=50, Entertainment=100, Other=200)
pie(budget,
    labels = as.character(budget),
    main="Monthly Budget",
    col=c("green","blue","yellow","magenta", "cyan", "orange"),
    border="black",
    clockwise=TRUE
)
```

```r
# from help(pie)
pie(rep(1, 24), col = rainbow(24), radius = 0.9)

pie.sales <- c(0.12, 0.3, 0.26, 0.16, 0.04, 0.12)
names(pie.sales) <- c("Blueberry", "Cherry",
    "Apple", "Boston Cream", "Other", "Vanilla Cream")
pie(pie.sales) # default colours
pie(pie.sales, col = c("purple", "violetred1", "green3",
                       "cornsilk", "cyan", "white"))
pie(pie.sales, col = gray(seq(0.4, 1.0, length = 6)))
pie(pie.sales, density = 10, angle = 15 + 10 * 1:6)
pie(pie.sales, clockwise = TRUE, main = "pie(*, clockwise = TRUE)")
segments(0, 0, 0, 1, col = "red", lwd = 2)
text(0, 1, "init.angle = 90", col = "red")

n <- 200
pie(rep(1, n), labels = "", col = rainbow(n), border = NA,
    main = "pie(*, labels=\"\", col=rainbow(n), border=NA,..")

## Another case showing pie() is rather fun than science:
## (original by FinalBackwardsGlance on http://imgur.com/gallery/wWrpU4X)
pie(c(Sky = 78, "Sunny side of pyramid" = 17, "Shady side of pyramid" = 5),
    init.angle = 315, col = c("deepskyblue", "yellow", "yellow3"), border = FALSE)

help(pie)
```
