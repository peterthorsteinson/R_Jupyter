## Mandelbrot Fractal

The Mandelbrot set is based on complex numbers: $$z = x + yi$$

The Mandelbrot set algorithm is:

1. Initiialize a set of complex numbers in the complex plane all set to zero:
$$z=0$$

2. Iterate the following statement where c is a set of complex numbers in the complex plane:
$$z_{n+1} = z_n^2 + c$$

3. The Mandelbrot set is the set where z remains bounded for all n.

Mandelbrot set images are generated by counting the number of iterations it takes for z to diverge for a given c, and
assign colors to each point in the complex plane according to the rate of divergence.

[Rosetta Code: Mandelbrot Set in R](https://rosettacode.org/wiki/Mandelbrot_set#R)  

```r
iterate.until.escape <- function(z, c, trans, cond, max=50, response=dwell) {
  #we iterate all active points in the same array operation,
  #and keeping track of which points are still iterating.
  active <- seq_along(z)
  dwell <- z
  dwell[] <- 0
  for (i in 1:max) {
    z[active] <- trans(z[active], c[active]);
    survived <- cond(z[active])
    dwell[active[!survived]] <- i
    active <- active[survived]
    if (length(active) == 0) break
  }
  eval(substitute(response))
}
 
re = seq(-2, 1, len=500)
im = seq(-1.5, 1.5, len=500)
c <- outer(re, im, function(x,y) complex(real=x, imaginary=y))
x <- iterate.until.escape(array(0, dim(c)), c,
                          function(z,c)z^2+c, function(z)abs(z) <= 2,
                          max=100)
image(x)
```

## Simple Moving Average

[Rosetta Code: Simple Moving Average in R](http://rosettacode.org/wiki/Averages/Simple_moving_average#R)  

```r
#concat concatenates the new values to the existing vector of values, then discards any values that are too old.
lastvalues <- local(
{
   values <- c(); 
   function(x, len)
   {
      values <<- c(values, x); 
      lenv <- length(values); 
      if(lenv > len) values <<- values[(len-lenv):-1]
      values
   }
})
 
#moving.average accepts a numeric scalars input (and optionally a length, i.e. the number of values to retain) and calculates the stateful moving average.
moving.average <- function(latestvalue, len=3)
{
   #Check that all inputs are numeric scalars
   is.numeric.scalar <- function(x) is.numeric(x) && length(x)==1L
   if(!is.numeric.scalar(latestvalue) || !is.numeric.scalar(len))
   {
      stop("all arguments must be numeric scalars")
   }
 
   #Calculate mean of variables so far  
   mean(lastvalues(latestvalue, len))
}
moving.average(5)  # 5
moving.average(1)  # 3
moving.average(-3) # 1
moving.average(8)  # 2
moving.average(7)  # 4

# try it out with some simulated time series data
x <- 1:200
y <- cos(x/40) + rnorm(200, sd=.2)
plot(x, y, type="l")
grid()
z <- lapply(y, moving.average)
lines(x, z, col="red")
```

## Bubble sort

[Wikipedia: Bubble sort](https://en.wikipedia.org/wiki/Bubble_sort)  
[Rosetta Code: Bubble sort in R](http://rosettacode.org/wiki/Sorting_algorithms/Bubble_sort#R)  

```r
bubblesort <- function(v) {
  itemCount <- length(v)
  repeat {
    hasChanged <- FALSE
    itemCount <- itemCount - 1
    for(i in 1:itemCount) {
      if ( v[i] > v[i+1] ) {
        t <- v[i]
        v[i] <- v[i+1]
        v[i+1] <- t
        hasChanged <- TRUE
      }
    }
    if ( !hasChanged ) break;
  }
  v
}
 
v <- c(9,8,7,3,1,100)
print(bubblesort(v))
```

## Roots of a quadratic function

[Wikipedia: Quadratic Equation](https://en.wikipedia.org/wiki/Quadratic_equation)  

A quadratic equation has the form:

$$ax^2 + bx + c = 0$$

The Quadratic Formula: For ax2 + bx + c = 0, the values of x which are the solutions of the equation are given by:

$$x = \frac{-b±\sqrt{b^2-4ac}}{2a}$$

[Rosetta Code: Quadratic Equation](http://rosettacode.org/wiki/Roots_of_a_quadratic_function#R)  

```r
quaddiscrroots <- function(a, b, c, tol=1e-5) {
  d <- b*b - 4*a*c + 0i
  root1 <- (-b + sqrt(d))/(2*a)
  root2 <- (-b - sqrt(d))/(2*a)
  if ( abs(Re(d)) < tol ) {
    list("real and equal", abs(root1), abs(root1))
  } else if ( Re(d) > 0 ) {
    list("real", Re(root1), Re(root2))
  } else {
    list("complex", root1, root2)
  }
}
 
for(coeffs in list(c(3,4,4/3), c(3,2,-1), c(3,2,1), c(1, -1e6, 1)) ) {
  cat(sprintf("roots of %gx^2 %+gx^1 %+g are\n", coeffs[1], coeffs[2], coeffs[3]))
  r <- quaddiscrroots(coeffs[1], coeffs[2], coeffs[3])
  cat(sprintf("  %s: %s, %s\n", r[[1]], r[[2]], r[[3]]))
}

curve(x^2+3*x-4, -10, 10, main="Roots of x^2+3x-4=0")
grid()
roots <- quaddiscrroots(1, 3, -4)
roots
abline(h=0, col="blue")
abline(v=roots[2], col="red")
abline(v=roots[3], col="red")
```

## Newton Method: Roots of a Function

$$ƒ(x) = ax^3 + bx^2 + cx$$ 

$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$

[Wikipedia: Newton's method](https://en.wikipedia.org/wiki/Newton%27s_method)  

```r
# newton method

newton <- function(f, tol=1E-12, x0=1, N=20) {
  h <- 0.001
  i <- 1; x1 <- x0
  p <- numeric(N)
  while (i <= N) {
    df.dx <- (f(x0+h)-f(x0))/h
    x1 <- (x0 - (f(x0)/df.dx))
    p[i] <- x1
    i <- i + 1
    if (abs(x1-x0) < tol) break
    x0 <- x1
  }
  return(p[1:(i-1)])
}

f <- function(x) { x^3 + 3*x^2 - 20*x - 8 }

x <- -10:10
    
p <- newton(f, x0=1, N=20)
p[length(p)]                # last one is best
plot(x, f(x), type='l', lwd=1.5, main="Newton Method")
abline(h=0)
abline(v=p[length(p)])
```

## The optimize Function

[The optimize Function](https://www.rdocumentation.org/packages/stats/versions/3.4.3/topics/optimize)

```r
# The optimize function searches interval from lower to upper for minimum/maximum of function wrt first argument
# one dimensional optimization
sales <- function(price) { 100 - 0.5 * price }
revenue <- function(price) { price * sales(price) }
par(mfrow=c(1, 2))
curve(sales, from=50, to=150, xname="price", ylab="Sales", main="Sales")
curve(revenue, from=50, to=150, xname="price", ylab="Revenue", main="Revenue")
optimum <- optimize(revenue, interval=c(50, 150), maximum=TRUE)
optimum$maximum
abline(v=optimum$maximum)
optimum$objective
abline(h=optimum$objective)
```
