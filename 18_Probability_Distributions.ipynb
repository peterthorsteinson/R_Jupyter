## Binomial Distribution

https://en.wikipedia.org/wiki/Binomial_distribution

The binomial distribution is the discrete probability distribution of the number of successes in a sequence of n independent Bernoulli trials. Each Bernoulli trial has a binary outcome of either yes with probability p, or no with probability q = 1 − p. A sequence of Bernoulli trial outcomes is called a Bernoulli process.


If the random variable X follows the binomial distribution with parameters n ∈ ℕ and p ∈ [0,1], then we write X ~ B(n, p). Then the probability of getting exactly k successes in n trials is given by the probability mass function:

\begin{equation*}
Pr(X=k)   = {n \choose k} p^k (1-p)^{ n-k}
\end{equation*}

Where:

-  k = 0, 1, 2, ... n
-  And the binomial coefficient (n choose k) a is defined as:

\begin{equation*}
{n \choose k}= \frac{n!}{k!(n-k)!}
\end{equation*}

The binomial coefficient ${n \choose k}$ represents the number of possible k-combinations.  

For example, lets say that you toss a fair coin with equal probability for heads and tails, 10 times. Then n=10 and p=0.5, and we have:
    
\begin{equation*}
Pr(X=k)   = {10 \choose k} 0.5^k 0.5^{10-k}
\end{equation*}


```r
# factorial
# N! = N*(N-1)*(N-2)*...(1)
# 4! = 4*3*2*1
factorial(4)
```

```r
# binomial distribution
Pr <- function(k, n, p) factorial(n)/(factorial(k)*factorial(n-k)) * p^k * (1-p)^(n-k)
k <- 0:10
names(k) = factor(k)
pr <- Pr(k, 10, 0.5)
barplot(pr, main="Binomial Distribution, n=10, p=0.5", xlab="k", ylab="Pr(X=k)", las=2)
```

```r
n <- 20
k <- seq(0, n, by = 1)
names(k) = k

# binomial distribution
pr <- dbinom(k, size=n, prob=0.5)
barplot(pr, main="Binomial Distribution, n=20, p=0.5", xlab="k", ylab="Pr(X=k)", las=2)

# binomial cumulative probability
cd <- pbinom(k, size=n, prob=0.5)
barplot(cd, main="Binomial Cumulative Probability, n=20, p=0.5", xlab="k", ylab="Cumulative Probability", las=2)
```

```r
# coin toss experiment

sample.space <- c(0,1)
p <- 0.5   # fair coin
n <- 20    # flip coin 20 times perh experiment
N <- 1000  # total number of experiments

counts <- rep(0, n + 1)
k <- 0:n
names(counts) = k

expr <- 1
while (expr < N) {
    flips <- sample(sample.space, size = n, replace = TRUE, prob = c(p, 1 - p))
    counts[sum(flips)+1] <- counts[sum(flips)+1]  + 1
    expr <- expr + 1
}
pr <- counts/N

barplot(pr, main="Coin Toss - Monte Carlo Simulation, n=20, p=0.5, N=1000", xlab="count", ylab="Pr(count)", las=2)
```

## Normal Distribution

The normal (Gaussian) distribution is defined as:

$$
f(x;\mu,\sigma^2) = \frac{1}{\sigma\sqrt{2\pi}} 
e^{ -\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2 }
$$

The **rnorm** function draws a random sample from a normal distribution.

The normal distribution has a bell shape.

```r
# normal distribution
library(ggplot2)
x <- rnorm(1000, 100, 15);
qplot(x)
```

## Poisson Distribution  

https://en.wikipedia.org/wiki/Poisson_distribution

The Poisson distribution is a discrete probability distribution that represents the probability of a given number
of events occurring in a fixed interval in space or time, where these events occur with a known average rate of occurance
that is independent of when or where the previous event occured.

\begin{equation*}
Pr(X=k) = \frac{\lambda^ke^{-\lambda}}{k!}
\end{equation*}

Where:
-  e is Euler's number (e=2.718...)
-  λ is the average number of events the given interval
-  k is the number of events occurring in the given interval

For examples: If there are, on average, 12 fatalities per day in a Boston, find the probability of 16 or more fatalities
in a given day. Therefore, λ = 12, and the probability is the area under the right tail of the distribution.


```r
# Assume 12 fatalities per day in a Boston, find the probability of 16 or more fatalities in a given day

# Probability of < 17 events per day is area under the lower tail of the probability density function
ppois(16, lambda=12)   # upper tail 


# Probability of >= 17 events per day is area under the upper tail of the probability density function
ppois(16, lambda=12, lower=FALSE)   # upper tail 
```

## Uniform Distribution

The uniform probability distribution is defined as:

$$f(x) = {\begin{cases} \frac{1}{b-a} a & \le x \le b \\ 0 & otherwise \\ \end{cases}}$$

The **runif** function draws n independent uniform random values.

```r
# uniform distribution
x <- runif(10, min=1, max=3)
names(x)=c(1:10)
barplot(x, main="runif(10, min=1, max=3)", xlab="samples")
```
