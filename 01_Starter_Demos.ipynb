## Using R Programmatically

````r
TRUE                     # TRUE
FALSE                    # FALSE
T                        # TRUE
F                        # FALSE
F + 0                    # 0
T + 0                    # 1
x <- c(T,F,T,F,F,F,T)
x                        # TRUE FALSE TRUE FALSE FALSE FALSE TRUE
as.integer(x)            # 1 0 1 0 0 0 1
sum(x)                   # 1+0+1+0+0+0+1 =  3
y <- c(1,2,3,4,5,6,7)
y                        # 1 2 3 4 5 6 7
sum(y)                   # 1+2+3+4+5+6+7 = 28
sum(y[x])                # 1+  3+      7 = 11
y %% 2                   # 1 0 1 0 1 0 1
sum(y %% 2)              # 1+0+1+0+1+0+1 =  4
y %% 2 == 0              # FALSE TRUE FALSE TRUE FALSE TRUE FALSE
as.integer(y %% 2 == 0)  # 0 1 0 1 0 1 0
sum(y %% 2 == 0)         # 0+1+0+1+0+1+0 =  3
as.integer(x %% 2 == 0)  # 0 1 0 1 1 1 0
y[x %% 2 == 0]           # 2 4 5 6
sum(y[x %% 2 == 0])      # 0+2+0+4+5+6+0 = 17 
````

````r
# get path for current working directory
wdsave <- getwd()
wdsave

# set working directory to specified path
setwd("/")

getwd()

setwd(wdsave)

getwd()
````

````r
# list all objects in environment
ls()            # display all objects in working environment
x <- 42         # create object x
ls()            # display all objects in working environment
y <- 3.141592   # create object y
ls()            # display all objects in working environment
exists("x")     # TRUE
rm(x)           # remove object x
exists("x")     # FALSE
rm(list = ls()) # remove all objects from working environment
ls()            # nothing displayed here!
````

````r
# console output
print("Hello World!")
"Hello World!"
````

````r
# plot a graph
x <- 1:5
y <- x^2
plot(x,y, type = "o")
````

````r
# Normal distribution (100000 observations)
x <- rnorm(100000)
hist(x, col='red', breaks=50, main="Normal Distribution (100000 observations)")
````

````r
# scatter plot: linear + noise 
x <- 1:10
y_intercept <- 2            # y_intercept
slope       <- 3            # slope
noise <- 2*rnorm(10)
y <- slope*x + y_intercept + noise
plot(x,y)
abline(y_intercept, slope)
````

````r
# Uniform distribution (100000 observations)"
x <- runif(100000)
hist(x, col='red', breaks=50, main="Uniform Distribution (100000 observations)")
````

## Scripting with R

-  Run script file at command prompt
-  Run script file within code

Given the following script file:
```
# test.R
myString <- "Hello, World!"
print(myString)
```

Run at command prompt and see output file test.Rout that is written:  
**R CMD BATCH test.R**

or

Run within code and see output returned:  
**source("test.R")**

````r
# run within code
source("scripts/test.R")
````

## Accessing documentation

-  R Help: **help()** and **?**
-  Vignettes and Demos: **browseVignettes()**, **vignette()** and **demo()**
-  Find objects by partial name: **apropos()**
-  Search Help System: **help.search()**
-  Search for keywords/phrases in docs: **RSiteSearch()**
-  Hypertext documentation: **help.start()**
-  Demonstrations of R Functionality: **demo()**
-  [CRAN Task Views](https://cran.r-project.org/web/views)
